```python
# Classes de base fournies
class Pets():
    def __init__(self, animals):
        self.animals = animals

    def walk(self):
        for animal in self.animals:
            print(animal.walk())

class Cat():
    is_lazy = True

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def walk(self):
        return f'{self.name} is just walking around'

class Bengal(Cat):
    def sing(self, sounds):
        return f'{sounds}'

class Chartreux(Cat):
    def sing(self, sounds):
        return f'{sounds}'

# Étape 1 : Créer la classe Siamese
class Siamese(Cat):
    def sing(self, sounds):
        return f'{sounds}'
    
    def meow(self):
        return f'{self.name} fait un miaulement distinctif de Siamois: Miaou!'

# Étape 2 : Créer une liste d'instances Cat
# Création des instances de chaque race
bengal_cat = Bengal("Tigrou", 3)
chartreux_cat = Chartreux("Grisou", 5)
siamese_cat = Siamese("Minou", 2)

# Liste contenant tous les chats
all_cats = [bengal_cat, chartreux_cat, siamese_cat]

# Étape 3 : Créer une instance Pets
sara_pets = Pets(all_cats)

# Étape 4 : Promenez les chats
print("Promenade des chats")
sara_pets.walk()

# Bonus : Démonstration des méthodes spécifiques
print("\nDémonstration des méthodes spécifiques ")
print(bengal_cat.sing("Roooar comme un tigre!"))
print(chartreux_cat.sing("Miaou discret"))
print(siamese_cat.sing("Miaou aigu et mélodieux"))
print(siamese_cat.meow())

# Démonstration de l'attribut de classe
print(f"\nAttribut de classe")
print(f"Est-ce que les chats sont paresseux ? {Cat.is_lazy}")
for cat in all_cats:
    print(f"{cat.name} est paresseux : {cat.is_lazy}")
```

```python
class Dog:
    def __init__(self, name, age, weight):
        """
        Initialise un chien avec un nom, un âge et un poids.
        
        Args:
            name (str): Le nom du chien
            age (int): L'âge du chien en années
            weight (float): Le poids du chien en kg
        """
        self.name = name
        self.age = age
        self.weight = weight
    
    def bark(self):
        """
        Fait aboyer le chien.
        
        Returns:
            str: Message d'aboiement du chien
        """
        return f"{self.name} c'est aboyer."
    
    def run_speed(self):
        """
        Calcule la vitesse de course du chien.
        
        Returns:
            float: La vitesse de course (poids / âge * 10)
        """
        return self.weight / self.age * 10
    
    def fight(self, other_dog):
        """
        Détermine le gagnant d'un combat entre deux chiens.
        Le calcul se base sur : vitesse_course * poids
        
        Args:
            other_dog (Dog): L'autre chien qui se bat
            
        Returns:
            str: Message indiquant quel chien a gagné
        """
        # Calculer la force de combat de chaque chien
        self_strength = self.run_speed() * self.weight
        other_strength = other_dog.run_speed() * other_dog.weight
        
        if self_strength > other_strength:
            return f"{self.name} a gagné le combat contre {other_dog.name}!"
        elif other_strength > self_strength:
            return f"{other_dog.name} a gagné le combat contre {self.name}!"
        else:
            return f"Match nul entre {self.name} et {other_dog.name}!"
    
    def __str__(self):
        """
        Représentation textuelle du chien.
        
        Returns:
            str: Description du chien
        """
        return f"{self.name} (âge: {self.age} ans, poids: {self.weight} kg)"


# Étape 2 : Créer des instances de chien
print("=== Création des chiens ===")
dog1 = Dog("Rex", 3, 25.5)
dog2 = Dog("Bella", 5, 18.0)
dog3 = Dog("Max", 2, 30.0)

print(f"Chien 1: {dog1}")
print(f"Chien 2: {dog2}")
print(f"Chien 3: {dog3}")

# Étape 3 : Tester les méthodes du chien
print("\n=== Test des méthodes ===")

# Test de la méthode bark()
print("\n--- Aboiements ---")
print(dog1.bark())
print(dog2.bark())
print(dog3.bark())

# Test de la méthode run_speed()
print("\n--- Vitesses de course ---")
print(f"Vitesse de {dog1.name}: {dog1.run_speed():.2f}")
print(f"Vitesse de {dog2.name}: {dog2.run_speed():.2f}")
print(f"Vitesse de {dog3.name}: {dog3.run_speed():.2f}")

# Test de la méthode fight()
print("\n--- Combats ---")
print(dog1.fight(dog2))
print(dog2.fight(dog3))
print(dog1.fight(dog3))

# Affichage des statistiques de combat pour comprendre les résultats
print("\n=== Statistiques de combat ===")
for dog in [dog1, dog2, dog3]:
    combat_strength = dog.run_speed() * dog.weight
    print(f"{dog.name}: Force = {combat_strength:.2f} (vitesse: {dog.run_speed():.2f} × poids: {dog.weight})")
```

```python
# Import de la classe Dog
class Dog:
    def __init__(self, name, age, weight):
        self.name = name
        self.age = age
        self.weight = weight
    
    def bark(self):
        return f"{self.name} says Woof!"

# Import du module random
import random

# Classe PetDog qui hérite de Dog
class PetDog(Dog):
    def __init__(self, name, age, weight):
        super().__init__(name, age, weight)
        self.trained = False
    
    def train(self):
        """Entraîne le chien - fait aboyer et marque comme entraîné"""
        print(self.bark())
        self.trained = True
    
    def play(self, *args):
        """Fait jouer le chien avec d'autres chiens"""
        # contient tous les noms des autres chiens
        print("They all play together!")
    
    def do_a_trick(self):
        """Fait faire un tour au chien s'il est entraîné"""
        if self.trained:
            tricks = ["does a barrel roll", "stands on his back legs", "shakes your hand", "plays dead"]
            print(f"{self.name} {random.choice(tricks)}")

# Étape 3 : Tests Petdog
if __name__ == "__main__":
    my_dog = PetDog("Fido", 2, 10)
    my_dog.train()
    my_dog.play("Buddy", "Max")
    my_dog.do_a_trick()
    
    print("\nTests supplémentaires")
    # Créer d'autres chiens pour plus de tests
    buddy = PetDog("Buddy", 3, 15)
    max_dog = PetDog("Max", 1, 8)
    
    # Entraîner les autres chiens
    buddy.train()
    max_dog.train()
    
    # Tester plusieurs tours aléatoires
    print("\nPlusieurs tours aléatoires:")
    for i in range(3):
        my_dog.do_a_trick()
    
    # Tester play avec différents arguments
    print("\nDifférentes façons de jouer:")
    buddy.play("Fido")
    max_dog.play("Fido", "Buddy", "Rex")
```

```python
class Person:
    def __init__(self, first_name, age):
        self.first_name = first_name
        self.age = age
        self.last_name = ""  # Initialise
    
    def is_18(self):
        """Retourne True si la personne a 18 ans ou plus, sinon False"""
        return self.age >= 18


class Family:
    def __init__(self, last_name):
        self.last_name = last_name
        self.members = []  # Liste vide pour stocker les objets Person
    
    def born(self, first_name, age):
        """Crée un nouvel objet Person et l'ajoute à la famille"""
        # Créer une nouvelle personne
        new_person = Person(first_name, age)
        # Attribuer le nom de famille
        new_person.last_name = self.last_name
        # Ajouter à la liste des membres
        self.members.append(new_person)
    
    def check_majority(self, first_name):
        """Vérifie si une personne peut sortir avec ses amis selon son âge"""
        # Rechercher la personne dans la liste des membres
        for member in self.members:
            if member.first_name == first_name:
                # Personne trouvée, vérifier son âge
                if member.is_18():
                    print("You are over 18, your parents Jane and John accept that you will go out with your friends")
                else:
                    print("Sorry, you are not allowed to go out with your friends.")
                return
        
        # Si la personne n'est pas trouvée
        print(f"Aucun membre de la famille ne s'appelle {first_name}")
    
    def family_presentation(self):
        """Affiche les informations de la famille"""
        print(f"Famille {self.last_name}")
        print("Membres de la famille :")
        for member in self.members:
            print(f"- {member.first_name}, {member.age} ans")


# Test des classes
if __name__ == "__main__":
    # Créer une famille
    famille_dupont = Family("Dupont")
    
    # Ajouter des membres à la famille
    famille_dupont.born("Marie", 20)
    famille_dupont.born("Pierre", 16)
    famille_dupont.born("Sophie", 25)
    famille_dupont.born("Lucas", 14)
    
    # Présenter la famille
    print("Présentation de la famille")
    famille_dupont.family_presentation()
    
    print("\nTests de majorité")
    # Tester la vérification de majorité
    famille_dupont.check_majority("Marie")    # Plus de 18 ans
    famille_dupont.check_majority("Pierre")   # Moins de 18 ans
    famille_dupont.check_majority("Sophie")   # Plus de 18 ans
    famille_dupont.check_majority("Lucas")    # Moins de 18 ans
    famille_dupont.check_majority("Jean")     # N'existe pas
    
    print("\nTest individuel d'une personne")
    # Test de la méthode is_18() directement
    personne_test = Person("Anna", 19)
    print(f"Anna a {personne_test.age} ans, majeure : {personne_test.is_18()}")
```
