Exercice 1

```python
class Cat:
    def __init__(self, cat_name, cat_age):
        self.name = cat_name
        self.age = cat_age

# Créons les objets Cat
cat1 = Cat("Minou", 3)
cat2 = Cat("Félix", 7)
cat3 = Cat("Garfield", 5)

# Créons une fonction pour trouver le chat le plus âgé
def find_oldest_cat(cat1, cat2, cat3):
    """
    Trouve le chat le plus âgé parmi trois chats.
    
    Args:
        cat1, cat2, cat3: Objets de la classe Cat
    
    Returns:
        Cat: L'objet chat le plus âgé
    """
    oldest_cat = cat1
    
    if cat2.age > oldest_cat.age:
        oldest_cat = cat2
    
    if cat3.age > oldest_cat.age:
        oldest_cat = cat3
    
    return oldest_cat

# Imprimons les détails du chat le plus âgé
oldest_cat = find_oldest_cat(cat1, cat2, cat3)
print(f"Le chat le plus âgé {oldest_cat.name} a {oldest_cat.age} années")

# Affichage supplémentaire pour vérification
print("\nDétails de tous les chats :")
print(f"- {cat1.name}: {cat1.age} ans")
print(f"- {cat2.name}: {cat2.age} ans") 
print(f"- {cat3.name}: {cat3.age} ans")
```

Exercice 2

```python
# Étape 1 : Créer la classe Dog
class Dog:
    def __init__(self, name, height):
        """
        Initialise un objet Dog avec un nom et une hauteur
        
        Args:
            name (str): Le nom du chien
            height (int/float): La hauteur du chien en cm
        """
        self.name = name
        self.height = height
    
    def bark(self):
        """Méthode pour faire aboyer le chien"""
        print(f"{self.name} goes woof !")
    
    def jump(self):
        """Méthode pour faire sauter le chien - saute à 2x sa hauteur"""
        jump_height = self.height * 2
        print(f"{self.name} sauts {jump_height} cm de haut !")

# Étape 2 : Créer des objets chien
davids_dog = Dog("Rex", 30)
sarahs_dog = Dog("Bella", 25)

# Étape 3 : Imprimer les détails du chien et appeler les méthodes
print("Détails des chiens")
print(f"Chien de David : {davids_dog.name}, hauteur : {davids_dog.height} cm")
print(f"Chien de Sarah : {sarahs_dog.name}, hauteur : {sarahs_dog.height} cm")

print("\nMéthodes bark()")
davids_dog.bark()
sarahs_dog.bark()

print("\nMéthodes jump()")
davids_dog.jump()
sarahs_dog.jump()

# Étape 4 : Comparer les tailles des chiens
print("\nComparaison des tailles")
if davids_dog.height > sarahs_dog.height:
    print(f"{davids_dog.name} ({davids_dog.height} cm) est plus grand que {sarahs_dog.name} ({sarahs_dog.height} cm)")
elif davids_dog.height < sarahs_dog.height:
    print(f"{sarahs_dog.name} ({sarahs_dog.height} cm) est plus grand que {davids_dog.name} ({davids_dog.height} cm)")
else:
    print(f"{davids_dog.name} et {sarahs_dog.name} ont la même taille ({davids_dog.height} cm)")

# Différence de taille
difference = abs(davids_dog.height - sarahs_dog.height)
print(f"Différence de taille : {difference} cm")
```

```python
class Song:
    """
    Classe pour représenter les paroles d'une chanson
    """
    
    def __init__(self, lyrics):
        """
        Initialise une nouvelle chanson avec ses paroles
        
        Args:
            lyrics (list): Liste des lignes de paroles
        """
        self.lyrics = lyrics
    
    def sing_me_a_song(self):
        """
        Affiche les paroles de la chanson ligne par ligne
        """
        for line in self.lyrics:
            print(line)
```

```python
class Zoo:
    def __init__(self, zoo_name):
        """
        Initialise un nouveau zoo avec un nom et une liste vide d'animaux.
        
        Args:
            zoo_name (str): Le nom du zoo
        """
        self.zoo_name = zoo_name
        self.animals = []
    
    def add_animal(self, new_animal):
        """
        Ajoute un nouvel animal à la liste des animaux.
        N'ajoute pas l'animal s'il est déjà présent.
        
        Args:
            new_animal (str): Le nom de l'animal à ajouter
        """
        if new_animal not in self.animals:
            self.animals.append(new_animal)
    
    def get_animals(self):
        """
        Affiche tous les animaux actuellement dans le zoo.
        """
        print(f"Animaux dans le zoo {self.zoo_name}:")
        for animal in self.animals:
            print(f"- {animal}")
    
    def sell_animal(self, animal_sold):
        """
        Retire un animal de la liste s'il existe.
        
        Args:
            animal_sold (str): Le nom de l'animal à vendre/retirer
        """
        if animal_sold in self.animals:
            self.animals.remove(animal_sold)
            print(f"{animal_sold} a été vendu et retiré du zoo.")
        else:
            print(f"{animal_sold} n'est pas dans le zoo.")
    
    def sort_animals(self):
        """
        Trie les animaux par ordre alphabétique et les groupe par première lettre.
        
        Returns:
            dict: Dictionnaire avec les lettres comme clés et listes d'animaux comme valeurs
        """
        # Trier les animaux par ordre alphabétique
        sorted_animals = sorted(self.animals)
        
        # Créer un dictionnaire pour grouper par première lettre
        grouped_animals = {}
        
        for animal in sorted_animals:
            first_letter = animal[0].upper()  # Première lettre en majuscule
            
            if first_letter not in grouped_animals:
                grouped_animals[first_letter] = []
            
            grouped_animals[first_letter].append(animal)
        
        return grouped_animals
    
    def get_groups(self):
        """
        Affiche les animaux groupés par première lettre.
        """
        grouped_animals = self.sort_animals()
        
        print("Animaux groupés par première lettre:")
        for letter, animals in grouped_animals.items():
            print(f"{letter}: {animals}")

if __name__ == "__main__":
    # Créons un zoo
    mon_zoo = Zoo("Zoo de Paris")
    
    # Ajoutons des animaux
    mon_zoo.add_animal("Lion")
    mon_zoo.add_animal("Zebra")
    mon_zoo.add_animal("Bear")
    mon_zoo.add_animal("Giraffe")
    mon_zoo.add_animal("Cat")
    mon_zoo.add_animal("Baboon")
    mon_zoo.add_animal("Cougar")
    
    # Essayons d'ajouter un animal déjà présent
    mon_zoo.add_animal("Lion")  # Ne sera pas ajouté
    
    # Affichons tous les animaux
    mon_zoo.get_animals()
    
    # Affichons les animaux groupés
    mon_zoo.get_groups()
    
    # Vendre un animal
    mon_zoo.sell_animal("Bear")
    
    # Afficher les groupes après la vente
    mon_zoo.get_groups()
```
