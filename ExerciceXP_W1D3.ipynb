Exercice 1

```python
class Cat:
    def __init__(self, cat_name, cat_age):
        self.name = cat_name
        self.age = cat_age
# Étape 1 : Création des objets Cat
cat1 = Cat("Max", 3)
cat2 = Cat("Félix", 7)
cat3 = Cat("Whiskers", 5)
# Étape 2 : Création d'une fonction pour trouver le chat le plus âgé
def find_oldest_cat(cat1, cat2, cat3):
    """
    Trouve le chat le plus âgé parmi trois chats donnés.
    
    Args:
        cat1, cat2, cat3: Objets de la classe Cat
    
    Returns:
        Cat: L'objet chat le plus âgé
    """
    oldest = cat1  # Commençons par supposer que le premier chat est le plus âgé
    
    if cat2.age > oldest.age:
        oldest = cat2
    
    if cat3.age > oldest.age:
        oldest = cat3
    
    return oldest
# Étape 3 : Imprimons les détails du chat le plus âgé
oldest_cat = find_oldest_cat(cat1, cat2, cat3)
print(f"Le chat le plus âgé est {oldest_cat.name} qui a {oldest_cat.age} ans.")
```

```python
# Étape 1 : Création de la classe Dog
class Dog:
    def __init__(self, name, height):
        """Initialise un chien avec un nom et une taille"""
        self.name = name
        self.height = height
    
    def bark(self):
        """Méthode pour faire aboyer le chien"""
        print(f"{self.name} goes woof !")
    
    def jump(self):
        """Méthode pour faire sauter le chien (hauteur * 2)"""
        jump_height = self.height * 2
        print(f"{self.name} sauts {jump_height} cm de haut !")
# Étape 2 : Créons des objets chien
davids_dog = Dog("Rex", 30)
sarahs_dog = Dog("Bella", 25)

print("=== DÉTAILS DES CHIENS ===")
# Étape 3 : Imprimons les détails et appeler les méthodes
print(f"Chien de David : {davids_dog.name}, Taille : {davids_dog.height} cm")
print(f"Chien de Sarah : {sarahs_dog.name}, Taille : {sarahs_dog.height} cm")

print("\n=== ACTIONS DES CHIENS ===")
# Appelons les méthodes bark() et jump() pour chaque chien
davids_dog.bark()
davids_dog.jump()

sarahs_dog.bark()
sarahs_dog.jump()

# Étape 4 : Comparons les tailles des chiens
print("\n=== COMPARAISON DES TAILLES ===")
if davids_dog.height > sarahs_dog.height:
    print(f"{davids_dog.name} est plus grand que {sarahs_dog.name}")
    print(f"Différence : {davids_dog.height - sarahs_dog.height} cm")
elif sarahs_dog.height > davids_dog.height:
    print(f"{sarahs_dog.name} est plus grand que {davids_dog.name}")
    print(f"Différence : {sarahs_dog.height - davids_dog.height} cm")
else:
    print(f"{davids_dog.name} et {sarahs_dog.name} font la même taille")

# Bonus : Créons d'autres chiens pour plus de démonstration
print("\n=== BONUS : AUTRES CHIENS ===")
max_dog = Dog("Max", 40)
lucy_dog = Dog("Lucy", 20)

print(f"Nouveau chien : {max_dog.name}, Taille : {max_dog.height} cm")
print(f"Nouveau chien : {lucy_dog.name}, Taille : {lucy_dog.height} cm")

max_dog.bark()
max_dog.jump()
lucy_dog.bark()
lucy_dog.jump()

# Trouver le plus grand chien
all_dogs = [davids_dog, sarahs_dog, max_dog, lucy_dog]
tallest_dog = max(all_dogs, key=lambda dog: dog.height)
print(f"\nLe plus grand chien est {tallest_dog.name} avec {tallest_dog.height} cm !")
```

```python
class Song:
    """Classe pour représenter les paroles d'une chanson"""
    
    def __init__(self, lyrics):
        """
        Initialise une nouvelle chanson avec ses paroles.
        
        Args:
            lyrics (list): Liste des lignes de paroles
        """
        self.lyrics = lyrics
    
    def sing_me_a_song(self):
        """Imprime chaque ligne des paroles sur une nouvelle ligne"""
        for line in self.lyrics:
            print(line)


# Exemple d'utilisation avec des paroles originales
example_song = Song([
    "Voici une chanson d'exemple",
    "Avec des paroles simples",
    "Pour tester notre classe",
    "Et voir comment elle fonctionne"
])

print("=== Démonstration de la classe Song ===")
print("Paroles de la chanson d'exemple :")
example_song.sing_me_a_song()

print("\n" + "="*40)

# Autre exemple
folk_song = Song([
    "Dans les montagnes lointaines",
    "Où chantent les oiseaux",
    "Une mélodie résonne",
    "Portée par le vent"
])

print("Paroles de la chanson folk :")
folk_song.sing_me_a_song()

print("\n" + "="*40)

# Démonstration avec une chanson plus courte
short_song = Song([
    "La la la",
    "Do ré mi",
    "Une petite mélodie"
])

print("Paroles de la petite chanson :")
short_song.sing_me_a_song()
```

```python
class Zoo:
    def __init__(self, zoo_name):
        """
        Initialise un nouveau zoo avec un nom et une liste vide d'animaux.
        
        Args:
            zoo_name (str): Le nom du zoo
        """
        self.zoo_name = zoo_name
        self.animals = []
    
    def add_animal(self, new_animal):
        """
        Ajoute un nouvel animal au zoo s'il n'y est pas déjà.
        
        Args:
            new_animal (str): Le nom de l'animal à ajouter
        """
        if new_animal not in self.animals:
            self.animals.append(new_animal)
            print(f"{new_animal} a été ajouté au {self.zoo_name}")
        else:
            print(f"{new_animal} est déjà dans le zoo")
    
    def get_animals(self):
        """
        Affiche tous les animaux actuellement dans le zoo.
        """
        if self.animals:
            print(f"Animaux dans le {self.zoo_name}:")
            for animal in self.animals:
                print(f"- {animal}")
        else:
            print(f"Aucun animal dans le {self.zoo_name}")
    
    def sell_animal(self, animal_sold):
        """
        Vend (retire) un animal du zoo s'il existe.
        
        Args:
            animal_sold (str): Le nom de l'animal à vendre
        """
        if animal_sold in self.animals:
            self.animals.remove(animal_sold)
            print(f"{animal_sold} a été vendu et retiré du {self.zoo_name}")
        else:
            print(f"{animal_sold} n'est pas dans le zoo")
    
    def sort_animals(self):
        """
        Trie les animaux par ordre alphabétique et les groupe par première lettre.
        
        Returns:
            dict: Dictionnaire avec les lettres comme clés et listes d'animaux comme valeurs
        """
        # Trier les animaux par ordre alphabétique
        sorted_animals = sorted(self.animals)
        
        # Créer un dictionnaire pour grouper par première lettre
        groups = {}
        
        for animal in sorted_animals:
            first_letter = animal[0].upper()
            if first_letter not in groups:
                groups[first_letter] = []
            groups[first_letter].append(animal)
        
        # Stocker le résultat pour l'utiliser dans get_groups()
        self.animal_groups = groups
        return groups
    
    def get_groups(self):
        """
        Affiche les animaux groupés par première lettre.
        """
        # S'assurer que les animaux sont triés et groupés
        if not hasattr(self, 'animal_groups'):
            self.sort_animals()
        
        if self.animal_groups:
            print(f"Animaux groupés dans le {self.zoo_name}:")
            for letter in sorted(self.animal_groups.keys()):
                print(f"{letter}: {self.animal_groups[letter]}")
        else:
            print(f"Aucun animal à grouper dans le {self.zoo_name}")


# Étape 2: Créer une instance de Zoo
ramat_gan_safari = Zoo("Ramat Gan Safari")

# Étape 3: Utiliser les méthodes du Zoo
print("=== Test de la classe Zoo ===\n")

# Ajouter des animaux
print("1. Ajout d'animaux:")
ramat_gan_safari.add_animal("Giraffe")
ramat_gan_safari.add_animal("Bear")
ramat_gan_safari.add_animal("Baboon")
ramat_gan_safari.add_animal("Cat")
ramat_gan_safari.add_animal("Cougar")
ramat_gan_safari.add_animal("Lion")
ramat_gan_safari.add_animal("Zebra")

# Essayer d'ajouter un animal déjà présent
ramat_gan_safari.add_animal("Lion")  # Ne devrait pas être ajouté

print("\n2. Affichage des animaux:")
ramat_gan_safari.get_animals()

print("\n3. Vente d'un animal:")
ramat_gan_safari.sell_animal("Bear")

print("\n4. Affichage après vente:")
ramat_gan_safari.get_animals()

print("\n5. Tri et groupement:")
groups = ramat_gan_safari.sort_animals()
print("Dictionnaire retourné par sort_animals():")
print(groups)

print("\n6. Affichage des groupes:")
ramat_gan_safari.get_groups()

print("\n=== Test avec un zoo vide ===")
zoo_vide = Zoo("Zoo Vide")
zoo_vide.get_animals()
zoo_vide.get_groups()
```
